```python
store_clustering = analyze_data[["store_id", "total_amount"]].groupby("store_id").\
    agg(["size", "mean", "median", "max", "min"])

store_clustering = store_clustering.reset_index(drop=True)
store_clustering
```

```python
import seaborn as sns

hexbin = sns.jointplot(x="mean", y="size", data=store_clustering["total_amount"], kind="hex")
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
store_clustering_sc = scaler.fit_transform(store_clustering)

kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit(store_clustering_sc)
store_clustering["cluster"] = clusters.labels_
store_clustering
```

```python
store_clustering.columns = ["月内件数", "月内平均値", "月内中央値", "月内最大値", "月内最小値", "cluster"]
store_clustering.groupby("cluster").count()
```

```python
store_clustering.groupby("cluster").mean()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(random_state=0)

x = tsne.fit_transform(store_clustering_sc)
tsne_df = pd.DataFrame(x)
tsne_df["cluster"] = store_clustering["cluster"]
tsne_df.columns = ["axis_0", "axis_1", "cluster"]
tsne_df
```

```python
tsne_df.plot(kind="scatter", x="axis_0", y="axis_1", c="cluster", cmap="viridis")
```

```python
sns.scatterplot(data=tsne_df, x="axis_0", y="axis_1", hue="cluster", style="cluster", sizes=(1000, 2000),
                palette="coolwarm")
```
