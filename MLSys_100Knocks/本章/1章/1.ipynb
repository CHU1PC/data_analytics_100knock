```python
import pandas as pd

m_store = pd.read_csv("m_store.csv", header=0)
m_store
```

```python
m_area = pd.read_csv("m_area.csv", header=0)
m_area
```

```python
tbl_order_4 = pd.read_csv("tbl_order_202004.csv", header=0)
tbl_order_4
```

```python
tbl_order_5 = pd.read_csv("tbl_order_202005.csv", header=0)
tbl_order_5
```

```python
order_all = pd.concat([tbl_order_4, tbl_order_5], ignore_index=True)
order_all
```

```python
import glob

tbl_files = glob.glob("tbl_order_*.csv")
```

```python
order_all = pd.DataFrame()

for file in tbl_files:
    order_all = pd.concat([order_all, pd.read_csv(file)], axis=0, ignore_index=True)
    print(file)
order_all
```

```python
order_all.isna().sum(axis=0)
```

```python
order_all.describe(include="all")
```

```python
order_all["order_accept_date"] = pd.to_datetime(order_all["order_accept_date"])
order_all["delivered_date"] = pd.to_datetime(order_all["delivered_date"])
order_all[["order_accept_date", "delivered_date"]].describe()
```

```python
order_all.describe(include="all")
```

```python
order_all = order_all[order_all["store_id"] != 999]
order_all.describe(include="all")
```

```python
order_data = pd.merge(order_all, m_store, on="store_id", how="inner")
order_data
```

```python
order_data = pd.merge(order_data, m_area, on="area_cd", how="inner")
order_data
```

```python
order_data.isna().sum()
```

```python
order_data.loc[order_data["takeout_flag"] == 0, "takeout_name"] = "デリバリー"
order_data.loc[order_data["takeout_flag"] == 1, "takeout_name"] = "お持ち帰り"
```

```python
order_data["takeout_name"] = order_data["takeout_flag"].map({0: "デリバリ", 1: "お持ち帰り"})
order_data
```

```python
order_data["status_name"] = order_data["status"].map({0: "受付", 1: "お支払い済み", 2: "お渡し済み"})
order_data
```
